**About Dataset**

This dataset contains 912 samples with 7 attributes. There are some missing values in this dataset.

Here are the columns in this dataset-



*   Brand: This indicates the manufacturer of the product i.e. Television.
*   Resolution: This has multiple categories and indicates the type of display i.e. LED, HD LED, etc.
*   Size: This indicates the screen size in inches.
*   Selling Price: This column has the Selling Price or the Discounted Price of the product
*   Original Price: This includes the Original Price of the product from the manufacturer.
*   Operating system: This categorical variable shows the type of OS like Android, Linux, etc.
*   Rating: Average customer ratings on a scale of 5.


**About The Project**

This dataset could be used to explore the current market scenario for Televisions. There are various types of screens with different operating systems offered by several manufacturers at competitive prices. Some questions this dataset could be used to answer are -


1.   To observe transaction data based on the mode (Discount vs. No Discount).
2.   Percentage Transaction Analysis Based on Market Needs (Promo).
3.   The Top 10 Brand by Transaction.
4.   Transaction Analysis Based on Market Needs.
5.   Transaction Analysis by Promo.
6.   Top 10 Brands by Transaction (Discount).
7.   Top 10 Brands by Transaction (No Discount).
8.   Transaction Analysis Based on Market Needs (Resolution).
9.   Percentage Transaction Analysis Based on Market Needs (Resolution).
10.  Percentage Transaction Analysis Based on Market Needs (Size).
11.  Distribution based on Size.




```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv('/kaggle/input/television-brands-ecommerce-dataset/TV_Final.csv')
data.head()
```

```python
data.info()
```

```python
data.rename(columns={'Size ': 'Size'}, inplace=True)
```

```python
#merubah data pada kolom brand menjadi proper text
data['Brand'] = data['Brand'].str.title()
data['Brand'].unique()
```

```python
#merubah data pada kolom Operating System dan Resolution menjadi upper text
data['Operating System'] = data['Operating System'].str.upper()
data['Resolution'] = data['Resolution'].str.upper()

data.head()
```

```python
# cek null values
cek_isna_brand = data['Brand'].isna().sum()
cek_isna_resolution = data['Resolution'].isna().sum()
cek_isna_os = data['Operating System'].isna().sum()

print(f"""
Cek null values di kolom brand            = {cek_isna_brand}
Cek null values di kolom Resolution       = {cek_isna_resolution}
Cek null values di kolom Operating System = {cek_isna_os}
""")
```

```python
missing_os = data['Operating System'].isna().sum()
missing_os
```

```python
# cek duplicated values
cek_duplicate_brand = data['Brand'].duplicated().sum()
cek_duplicate_resolution = data['Resolution'].duplicated().sum()
cek_duplicate_os = data['Operating System'].duplicated().sum()

print(f"""
Cek duplicated values di kolom brand            = {cek_isna_brand}
Cek duplicated values di kolom Resolution       = {cek_isna_resolution}
Cek duplicated values di kolom Operating System = {cek_isna_os}
""")
```

```python
# cek duplicated di kolom operating system
duplicated_os = data[data['Operating System'].duplicated()]

duplicated_os
```

```python
#selling price adalah harga sesudah discount

# Fungsi untuk menentukan nilai kolom "discount"
def calculate_discount(row):
    if row['Selling Price'] == row['Original Price']:
        return "No Discount"
    else:
        return "Discount"

# Menerapkan fungsi ke setiap baris dan membuat kolom "discount" baru
data['Promo'] = data.apply(calculate_discount, axis = 1)

data.head()
```

```python
# cek data yang menggunakan promo Discount
no_discount = data[data['Promo'] == 'Discount']
no_discount
```

```python
# cek data yang menggunakan promo Discount
no_discount = data[data['Promo'] == 'No Discount']
no_discount
```

```python
# menambahkan kolom Transaction
def transaction (row):
    if row['Promo'] == 'Discount':
        return 1
    else:
        return 1

data['Transaction'] = data.apply(transaction, axis = 1)
data.head()
```

```python
discount = data[data['Promo'] == 'Discount']['Transaction'].sum()
nodiscount = data[data['Promo']=='No Discount']['Transaction'].sum()
peningkatan= ((discount-nodiscount)/nodiscount) * 100

print('Total Transaksi yang menggunakan discount : ', discount)
print('Total Transaksi yang tidak menggunakan discount : ', nodiscount)



print(f'Dengan demikian, setelah ada promo Discount, peningkatan penjualan mencapai puncaknya dengan pertumbuhan yang signifikan sebesar {peningkatan:.2f}%')
```

```python
labels = ['Discount','No Discount']
values = [discount,nodiscount]

plt.figure(figsize = (5,6))
plt.bar(labels, values, color='darkcyan')
plt.xlabel('Mode')
plt.ylabel('Total Transaction');
plt.title(f"""Discount vs No Discount
+ {peningkatan:.2f}%""")

for x,y in enumerate(values):
    plt.text(x, y , str(y), ha='center', va='bottom')
plt.show()
```

```python
plt.pie(values, labels=labels, autopct='%1.1f%%', startangle =90, colors=sns.color_palette('Set2'))
plt.title('Percentage Transaction Analysis Based on Market Needs (Promo)')

plt.show()
```

```python
# Mencari 10 brand terlaris
brand = data.groupby(data['Brand'])['Transaction'].sum()
brand = brand.sort_values(ascending = False)
top10 = brand.head(10)

# buat barh
plt.figure(figsize=(8,10))
top10 = top10.sort_values(ascending=True)
top10.plot(kind='barh',color = "darkcyan")


plt.xlabel('Total Transaction')
plt.title("The Top 10 Brand by Transaction")

for x, y in enumerate(top10):
    plt.text(y-8,x, str(y), ha = 'left',va = 'center', color='white')

plt.show()


```

```python
#Brands and Resolution

sns.set_style('white')
ax = sns.countplot(y="Brand", hue="Resolution", data=data, order=data["Brand"].value_counts().iloc[:5].index)
plt.title('Transaction Analysis Based on Market Needs')
plt.xlabel('Total Transaction')
# Menampilkan jumlah data pada setiap bar
for p in ax.patches:
    if p.get_width() != 0.0:
        ax.annotate(f'{p.get_width()}', (p.get_x() + p.get_width(), p.get_y()), ha='left', va='top')


# hue="Resolution": Ini mengatur parameter yang akan diplotkan dengan menggunakan warna yang berbeda.
# iloc[:5] : mengambil 5 brand teratas yang kemunculannya paling banyak,
# dan index mengambil indeks dari 5 brand tersebut.
```

```python

sns.set_style('white')
ax = sns.countplot(y="Brand", hue="Promo", data=data, order=data["Brand"].value_counts().iloc[:5].index)
plt.xlabel('Total Transaction')
plt.title('Transaction Analysis by Promo')

# Menampilkan jumlah data pada setiap bar
for p in ax.patches:
    if p.get_width() != 0.0:
        ax.annotate(f'{p.get_width()}', (p.get_x() + p.get_width(), p.get_y()), ha='left', va='top')
```

```python
brand_d = data[data['Promo'] == 'Discount'].groupby(data['Brand'])['Transaction'].sum()
brand_n = data[data['Promo'] == 'No Discount'].groupby(data['Brand'])['Transaction'].sum()

brand_d = brand_d.sort_values(ascending =False).head(10)
brand_n = brand_n.sort_values(ascending =False).head(10)

print('-----Transaction by Brand (Discount)-----')
print(brand_d)
print('\n-----Transaction by Brand (No Discount)-----')
print(brand_n)
```

```python

# Membuat figur dan sumbu pertama
fig, axs = plt.subplots(1, 2, figsize=(16, 8))  # 1 baris dan 2 kolom

# Data untuk plot pertama
brand_d = brand_d.sort_values(ascending=True)

# Plot pertama
brand_d.plot(kind='barh', color='royalblue', ax=axs[0])
axs[0].set_xlabel('Total Transaction')
axs[0].set_title('Top 10 Brands by Transaction (Discount)')

for x, y in enumerate(brand_d):
    axs[0].text(y - 8, x, str(y), ha='left', va='center', color='white')

# Data untuk plot kedua
brand_n = brand_n.sort_values(ascending=True)

# Plot kedua
brand_n.plot(kind='barh', color='darkorange', ax=axs[1])
axs[1].set_xlabel('Total Transaction')
axs[1].set_title('Top 10 Brands by Transaction (No Discount)')

for x, y in enumerate(brand_n):
    axs[1].text(y - 2, x, str(y), ha='left', va='center', color='white')

plt.tight_layout()  # Untuk memastikan tata letak yang rapi
plt.show()
```

```python
label = ['ULTRA HD LED', 'HD LED', 'FULL HD LED', 'QLED ULTRA HD','HD PLASMA']
resolution = data['Resolution'].value_counts().sort_values(ascending=True)


plt.figure(figsize=(9,7))
resolution.plot(kind='barh', color='darkcyan')
plt.xlabel('Total Transaction')
plt.title('Transaction Analysis Based on Market Needs (Resolution)')

for x,y in enumerate(resolution):
    plt.text(y-3,x, str(y), ha='right', va='center', color='white')


```

```python
# pie plot
plt.pie(resolution, labels=label,  autopct='%1.1f%%', colors = sns.color_palette(('Set2')));
plt.title('Percentage Transaction Analysis Based on Market Needs (Resolution)')
plt.show();
```
